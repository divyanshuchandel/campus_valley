```python
def add_employee():
    name = input("Enter employee name: ")
    age = input("Enter employee age: ")
    salary = input("Enter employee salary: ")

    with open("EmpData.txt", "a") as file:
        file.write(f"Name: {name}, Age: {age}, Salary: {salary}\n")
    print("Employee details added successfully!")


def main():
    while True:
        print("\nOptions:")
        print("1. Add employee details")
        print("2. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_employee()
        elif choice == "2":
            print("Exiting the program...")
            break
        else:
            print("Invalid choice. Please enter either '1' or '2'.")


if __name__ == "__main__":
    main()
```

```python
def calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            return "Error! Division by zero is not allowed."
        else:
            return num1 / num2
    else:
        return "Invalid operator"


def main():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    op = input("Enter an arithmetic operator (+, -, *, /): ")

    result = calculator(num1, num2, op)
    print("Result:", result)


if __name__ == "__main__":
    main()
```

```python
def division(num1, num2):
    try:
        result = num1 / num2
        return result
    except ZeroDivisionError:
        return "Error! Division by zero is not allowed."


def main():
    try:
        num1 = float(input("Enter the numerator: "))
        num2 = float(input("Enter the denominator: "))
        result = division(num1, num2)
        print("Result:", result)
    except ValueError:
        print("Error! Please enter valid numbers.")


if __name__ == "__main__":
    main()
```

```python
def factorial(num):
    if num < 0:
        return "Factorial is not defined for negative numbers."
    elif num == 0:
        return 1
    else:
        fact = 1
        for i in range(1, num + 1):
            fact *= i
        return fact


def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True


def is_palindrome(string):
    return string == string[::-1]


def is_armstrong(num):
    order = len(str(num))
    temp = num
    sum = 0
    while temp > 0:
        digit = temp % 10
        sum += digit**order
        temp //= 10
    return num == sum


def main():
    while True:
        print("\nOptions:")
        print("1. Calculate the factorial of a number")
        print("2. Check whether a given number is prime or not")
        print("3. Check if a given string is a palindrome")
        print("4. Check if a given number is Armstrong or not")
        print("5. Exit the program")

        choice = input("Enter your choice: ")

        if choice == "1":
            num = int(input("Enter a number to calculate its factorial: "))
            print("Factorial of", num, "is", factorial(num))
        elif choice == "2":
            num = int(input("Enter a number to check if it's prime: "))
            print(num, "is prime." if is_prime(num) else num, "is not prime.")
        elif choice == "3":
            string = input("Enter a string to check if it's a palindrome: ")
            print(
                string,
                "is a palindrome." if is_palindrome(string) else string,
                "is not a palindrome.",
            )
        elif choice == "4":
            num = int(input("Enter a number to check if it's Armstrong: "))
            print(
                num, "is Armstrong." if is_armstrong(num) else num, "is not Armstrong."
            )
        elif choice == "5":
            print("Exiting the program...")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")


if __name__ == "__main__":
    main()
```

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less_than_pivot = [x for x in arr[1:] if x <= pivot]
        greater_than_pivot = [x for x in arr[1:] if x > pivot]
        return quicksort(less_than_pivot) + [pivot] + quicksort(greater_than_pivot)


def main():
    user_input = input("Enter space-separated integers to be sorted: ")
    unsorted_array = list(map(int, user_input.split()))
    sorted_array = quicksort(unsorted_array)
    print("Sorted array:", sorted_array)


if __name__ == "__main__":
    main()
```
